## Spin Orbit Coupling (SOC)
SOC is the physical interaction between an electron's orbital motion about a nucleus and its spin.
  - Also known as *Russel-Saunders coupling* for multi-electron systems.

SOC causes energy levels to split into multiple, closely spaced levels, which only differ by a small amount of energy.  
  - Splitting of these energy levels is caused by the total angular momentum quantum number J, which takes values of
  $$ |L + S| ..to.. |L - S| $$
   where L is the total orbital angular quantum number, and S is the total instinstic spin quantum number in mutli-electron systems.

L and S are obtained by taking the sum of all the orbitals the electrons reside in (mₗ), and the sum of their spins (mₛ).
  - In single-electron systems, these are denoted as l and s for simplicity.
  - J is instead denoted j,  which also reflects a single electron system.

## SOC: Perturbation Theory
Given a solvable Hamiltonian with known eigenfunctions and eigenvalues for a simple, solveable system,
$$ \hat{H} Ψ = EΨ  $$

we can utilize $$ \hat{H} $$ such that for a different Hamiltonian that we initally cannot solve: $$ {H}̕ $$

we can write it to approximately solve it as:
$$ {H}̕ = \hat{H} + \hat{V} = \hat{H} + λH +... \tag{1}$$

where H is the order of perturbation for a given system, and λ is the magnitude of the perturbation.

Since we know both the eignfunctions and eigenvalues for a one-electron system, we can directly solve for the energy, seen in equation (2).

## SOC: Relativistic Effects
Relativistic effects occur when an atom's electron approaches the speed of light.
  - As a electron moves through the nucleus' electric field, this field, at high velocities, transforms into an effective magnetic field (via relativistic motion - meaning classical mechanics no longer apply).
  - This magnetic field interacts with our electron's spin and produces spin orbit coupling.

As an electron's velocity increases, so do relativistic effects, which increases SOC interactions.
  - Very common in heavy elements as they have more protons than light elements, resulting in higher Coulomb interactions due to increased nuclear charges.
  - A higher Coulombic charge pulls electrons closer to the nucleus, increasing electron density about the nucleus, which in turn increases electron velocities.

Since we are dealing with one-electron systems, relativistic effects will not be as prominent and are generally ignored.

## SOC: The Spin Orbit (Relativistic) Correction

Since we know the eignfunctions associated for a one-electron system, we can directly calculate the energy associated with the SOC.

The energy will be composed of two parts $E = E_0 + Eₛₒ \tag{2}$ where
$$ E_0 = -Z²/2n² $$
and $ Eₛₒ $, which under first order perturbation theory is
$$ Eₛₒ = \frac{Ze²} {8πϵₒm²c² }\frac{1}{r³}{𝐋·S} $$
which after expanding both radial and angular parts, appears as equation (3), seen below.

Note: the angular component uses J² instead of J as J² does commute with L² and S², but J does not commute with L and S.
  - This is because we cannot simultaneously know all components of J (ex. [Jₓ, Jᵧ] ≠ 0)


### Calculating the SOC Constant
The simplified spin-orbit coupling constant, $\zeta$, for a hydrogenic atom is given by:

$$ \zeta = \frac{Z^4 e^2 \hbar^2}{4\pi\epsilon_0^2 m_e^2 c^2 n^3 l(l+1)(l+1/2)} \tag{3}$$

This equation is specifically used for single-electron systems and assumes it is being used for hydrogen or hydrogen-like atoms (such as He+, Li²+).

```python
import scipy.constants as const

# Everytime we have a definition we have to state any parameters/variables associated with the definition.
  # We repeat this for each definition, even if we have previously stated a variable to be consistent.

def calculate_zeta(Z, n, l):
  """
  Calculates the spin-orbit coupling constant (zeta) for a hydrogenic atom.

  Parameters
  ----------
  Z : int
      The atomic number.
  n : int
      The principal quantum number.
  l : int
      The orbital angular momentum quantum number.

  Returns
  -------
  float
      The spin-orbit coupling constant (zeta).
  """
  # Defining physical constants (in SI units).
  e = const.e
  hbar = const.hbar
  eps0 = const.epsilon_0
  me = const.electron_mass
  c = const.c

  # Calculating Zeta
  numerator = (Z**4) * (e**2) * (hbar**2)
  denominator = 4 * const.pi * (eps0**2) * (me**2) * (c**2) * (n**3) * l * (l + 1) * (l + 0.5)

  # == is a equality operator and it essentially states that if l= 0, Zeta must equal 0 as l is in the denominator.
    # This makes sense as at low Z, relativistic effects are not as apparent.
  if l == 0:
      return 0.0 # Spin-orbit coupling is typically zero for s-orbitals (since l=0).

  zeta = numerator / denominator
  return zeta

# Here we can choose numerical values to be substituted into our formula, with minimal limitations.
  # The example here will represent a Helium cation, with its electron located in a 2p orbital.
Z_val_example = 2
n_val_example = 3
l_val_example = 1

# We notice that when we increase our Z value to a higher number, zeta gets larger.

calculated_zeta = calculate_zeta(Z_val_example, n_val_example, l_val_example)
print(f"Calculated zeta for Z={Z_val_example}, n={n_val_example}, l={l_val_example}: {calculated_zeta}")
```

### Calculating the Spin-Orbit Corrected Energy

We can use `total_energy` to calculate the energy of a system, including the spin-orbit correction. `E0` represents the unperturbed energy, and `zeta` will represent our SOC. Since our SOC for a one-electron system is very small, to see how it would apply, we will use a ficticous value.

```python
import numpy as np

def E0(n, Z):
  """
  Calculates the unperturbed energy of a hydrogenic atom.

  Parameters
  ----------
  n : int
      The principal quantum number.
  Z : int
      The atomic number of an atom.

  Returns
  -------
  float
      The unperturbed energy E0.
  """

  E0 = -Z**2 / (2 * n**2) #This calculates our unperturbed energy.
  return E0

def deltaE_SO(l, j, zeta):
  """
  Calculates the spin-orbit correction term.

  Parameters
  ----------
  l : int
      The orbital angular momentum quantum number.
  j : float
      The total angular momentum quantum number.
  zeta : float
      The spin-orbit coupling constant.

  Returns
  -------
  float
      The spin-orbit correction energy.
  """

  s = 0.50
  LS = 0.5 * (j*(j+1) - l*(l+1) - s*(s+1))
  return zeta * LS # This is another way to calculate our SOC term that includes j.

def total_energy(n, l, j, Z, zeta):
  """
  Calculates the total energy including the spin-orbit correction.

  Parameters
  ----------
  n : int
      The principal quantum number.
  l : int
      The orbital angular momentum quantum number.
  j : float
      The total angular momentum quantum number.
  Z : int
      The atomic number.
  zeta : float
      The spin-orbit coupling constant.

  Returns
  -------
  float
      The total energy with spin-orbit correction.
  """
  if not isinstance(l, int) or l < 0:
    raise ValueError("The orbital angular momentum quantum number (l) must be a non-negative integer.")
  if l >= n:
    raise ValueError("The orbital angular momentum quantum number (l) must be less than the principal quantum number (n).")
  return E0(n, Z) + deltaE_SO(l, j, zeta) # This adds our unperterbed and perterbed energy together.
```

```python
# Using our Helium example from before:
  # Set n=3, l=1 (p-orbital) state.
  # For a single electron, s = 0.5.
# J can take values |l + s| to |l - s|.
  # J can be 1.5 or 0.5.

# Define the parameters
n_val = 3
l_val = 1
Z_val = 2
zeta_val = 0.1 # Since our zeta value is small for a one-electron system, substitute a higher value number to see results.

# Calculate for J = 1.5
j_val_1 = 1.5
total_energy_j1_5 = total_energy(n_val, l_val, j_val_1, Z_val, zeta_val)
print(f"Total Energy for n={n_val}, l={l_val}, J={j_val_1}, Z={Z_val}: {total_energy_j1_5}")

# Calculate for J = 0.5
j_val_2 = 0.5
total_energy_j0_5 = total_energy(n_val, l_val, j_val_2, Z_val, zeta_val)
print(f"Total Energy for n={n_val}, l={l_val}, J={j_val_2}, Z={Z_val}: {total_energy_j0_5}")

# You can see the difference in energy due to the spin-orbit coupling for different J values.
```
